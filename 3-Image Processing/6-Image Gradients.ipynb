## Sobel Derivatives

```python
import cv2
import matplotlib.pyplot as plt

filepath = "C:\\Users\\Aung Thiha\\Computer Vision with Python\\Image Data Sets\\morpho.png"

original_image = cv2.imread(filepath)

gray = cv2.cvtColor(original_image,cv2.COLOR_BGR2GRAY)

sobelX = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=3)
sobelY = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=3)
sobel = cv2.Sobel(gray,cv2.CV_64F,1,1,ksize=3)

plt.subplot(131),plt.imshow(sobelX),plt.title("Sobel X")
plt.subplot(132),plt.imshow(sobelY),plt.title("Sobel Y")
plt.subplot(133),plt.imshow(sobel),plt.title("Sobel")
plt.show()

cv2.imshow("Sobel X",sobelX)
cv2.imshow("Sobel Y",sobelY)
cv2.imshow("Sobel Y",sobel)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Laplacian Derivatives
