# Simple Thresholding and Otsu Thresholding

```python
import cv2
import matplotlib.pyplot as plt

filepath = "C:\\Users\\Aung Thiha\\Computer Vision with Python\\Image Data Sets\\bookpage.jpg"
img = cv2.imread(filepath)

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

retval,threshold1 = cv2.threshold(gray,10,255,cv2.THRESH_BINARY)
retval,threshold2 = cv2.threshold(gray,10,255,cv2.THRESH_BINARY_INV)
retval,threshold3 = cv2.threshold(gray,10,255,cv2.THRESH_TOZERO)
retval,threshold4 = cv2.threshold(gray,10,255,cv2.THRESH_TOZERO_INV)
retval,threshold5 = cv2.threshold(gray,10,255,cv2.THRESH_TRUNC)
retval,threshold6 = cv2.threshold(gray,10,255,cv2.THRESH_OTSU)


plt.subplot(2,4,1),plt.imshow(img),plt.title("Original")
plt.subplot(2,4,2),plt.imshow(gray),plt.title("Gray")
plt.subplot(2,4,3),plt.imshow(threshold1),plt.title("BINARY")
plt.subplot(2,4,4),plt.imshow(threshold2),plt.title("BINARY_INV")
plt.subplot(2,4,5),plt.imshow(threshold3),plt.title("TOZERO")
plt.subplot(2,4,6),plt.imshow(threshold4),plt.title("TOZERO_INV")
plt.subplot(2,4,7),plt.imshow(threshold5),plt.title("TRUNC")
plt.subplot(2,4,8),plt.imshow(threshold6),plt.title("OTSU")

cv2.imshow("ath",img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Adaptive Thresholding 

```python
import matplotlib.pyplot as plt

filepath = "C:\\Users\\Aung Thiha\\Computer Vision with Python\\Image Data Sets\\bookpage.jpg"
img = cv2.imread(filepath)

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

threshold1 = cv2.adaptiveThreshold(gray , 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 3)
threshold2 = cv2.adaptiveThreshold(gray , 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 115, 3) 

plt.subplot(1,2,1)
plt.imshow(threshold1)
plt.title("Adaptive One")

plt.subplot(1,2,2)
plt.imshow(threshold2)
plt.title("Adaptive Two")
```
