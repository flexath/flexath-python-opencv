```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

filepath = "C:\\Users\\Aung Thiha\\Pictures\\aungthiha.jpg"

img = cv2.imread(filepath)

# img.shape returns (height,width,channel(color))
print("original dimension : ",img.shape)

resize = cv2.resize(img,dsize=None,fx=0.6,fy=0.5)
print("New dimension : ",resize.shape)

plt.xlabel("Width")
plt.ylabel("Height")
plt.imshow(resize)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

filepath = "C:\\Users\\Aung Thiha\\Pictures\\aungthiha.jpg"

img = cv2.imread(filepath)

# img.shape returns (height,width,channel(color))
print("original dimension : ",img.shape)

# 30 precent of image
width = int(img.shape[1]*(30/100))

# 50 precent of image
height = int(img.shape[0]*(50/100))

desired_size = (width,height)

resize_image = cv2.resize(img,dsize=desired_size)
print("New dimension : ",resize_image.shape)

plt.xlabel("Width")
plt.ylabel("Height")
plt.imshow(resize_image)
```

### crop image

# Translation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('C:\\Users\\Aung Thiha\\Pictures\\aungthiha.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,100]])
dst = cv2.warpAffine(img,M,(cols,rows))
print(M)
plt.imshow(dst)
cv2.imshow('img',dst)

cv2.waitKey(0)
cv2.destroyAllWindows()
plt.show()
```

# Rotation
    
https://www.tutorialkart.com/opencv/python/opencv-python-rotate-image/


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filepath = "C:\\Users\\Aung Thiha\\Pictures\\aungthiha.jpg"

img = cv2.imread(filepath)

width = (img.shape[1])/2
height = (img.shape[0])/2
center = (width,height)
angle = 320

desired_width = img.shape[1]
desired_height = img.shape[0]
desired_size = (desired_width,desired_height)

desired_size = (desired_width,desired_height)
scale = 0.6

#getRotationMatrix2D(center, angle, scale)
M = cv2.getRotationMatrix2D(center,angle,scale)

#cv2.warpAffine(img, M, (width, height))
rotate = cv2.warpAffine(img,M,desired_size)

plt.imshow(rotate)
```

# Affine Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filepath = "C:\\Users\\Aung Thiha\\Pictures\\aungthiha.jpg"
img = cv2.imread(filepath)

cols,rows,channel = img.shape
rows = int(rows/2)
cols = int(cols/2)
center = (rows,cols)

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[100,100],[200,50],[100,250]])

#M = np.float32([[1,0,200],[0,1,100]])
b = cv2.getAffineTransform(pts1,pts2)
a = cv2.warpAffine(img,b,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(a),plt.title('Output')
```

# Perspective Transformation ( like zooming)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filepath = "C:\\Users\\Aung Thiha\\Pictures\\aungthiha.jpg"
img = cv2.imread(filepath)

cols,rows,channel = img.shape
rows = int(rows/2)
cols = int(cols/2)
center = (rows,cols)

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

#M = np.float32([[1,0,200],[0,1,100]])
b = cv2.getPerspectiveTransform(pts1,pts2)
a = cv2.warpPerspective(img,b,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(a),plt.title('Output')
```
