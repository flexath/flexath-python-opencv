# Hough Circle Transform with Images

https://www.youtube.com/watch?v=Ltqt24SQQoI

https://docs.opencv.org/3.4/d4/d70/tutorial_hough_circle.html

```python
import cv2
import numpy as np

cv2.namedWindow("Hough Circle Transform" , cv2.WINDOW_NORMAL)
img = cv2.imread("C:\\Users\\Aung Thiha\\Computer Vision with Python\\Image Data Sets\\circle.jpg")
img = img.copy()
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur = cv2.medianBlur(gray,5)

circles = cv2.HoughCircles(blur,cv2.HOUGH_GRADIENT,1,100,param1=180,param2=50,minRadius=0,maxRadius=0)

#detected_circles = np.uint16(np.around(circles))
for (x_center,y_center,radius) in circles[0,:]:
    cv2.circle(img,(x_center,y_center),radius,(125,255,125),5)
    cv2.circle(img,(x_center,y_center),5,(0,0,0),3)

cv2.imshow("Hough Circle Transform" , img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Hough Circle Transform with Live Webcam

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while cap.isOpened() is True:
    cv2.namedWindow("Hough Circle Transform" , cv2.WINDOW_NORMAL)
    ret,frame = cap.read()
    copy = frame.copy()
    gray = cv2.cvtColor(copy,cv2.COLOR_BGR2GRAY)
    
    blur = cv2.medianBlur(gray,15)
    #blur = cv2.Canny(blur,20,150)
    circles = cv2.HoughCircles(blur,cv2.HOUGH_GRADIENT,1,150,param1=180,param2=50,minRadius=0,maxRadius=0)
    
    #detected_circles = np.uint16(np.around(circles))
    if circles is not None:
        for (x_center,y_center,radius) in circles[0,:]:
            cv2.circle(gray,(x_center,y_center) , radius , (0,0,0) , 5)
            cv2.circle(gray,(x_center,y_center) , 5 , (0,0,0) , 5)
            if circles is None:
                continue
    else:
        print("Wrong")
    
    cv2.imshow("Hough Circle Transform" , gray)
    key = cv2.waitKey(1)
    if key == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
